### Binds all motivation columns to a binary system, away from bool (for easier data processing and mean computes), checking "Not Sure" cells as 0 again:

```python
motivation_cols = [c for c in df.columns if c.startswith("motivation_")] #Finding Relevant Labels to Parse Thru
df[motivation_cols] = (
    df[motivation_cols]
      .replace({"Yes": 1, "No": 0, "Not Sure": 0}) #bind binaries
)
```

```python
df.head()
```

### Binds all motivation columns to a binary system, away from bool (for easier data processing and mean computes), checking "Not Sure" cells as 0 again:

```python
event_cols = [c for c in df.columns if c.startswith("event_")]

df[event_cols] = (
    df[event_cols]
      .replace({"Yes": 1, "No": 0, "Not Sure": 0})
)
```

```python
df.head()
```

binds Yes to 1, No to 0, fills Not-Sure again as 0 (unclear resp):

```python
df[["hear_tv", "hear_radio"]] = (
    df[["hear_tv", "hear_radio"]]
      .replace({"Yes": 1, "No": 0, "Not Sure": 0})
)
```

### Binds all interest columns to a binary system, away from bool

```python
interest_cols = [c for c in df.columns if c.startswith("interest_")]

df[interest_cols] = (
    df[interest_cols]
      .replace({"Yes": 1, "No": 0, "Not Sure": 0})
)
```
